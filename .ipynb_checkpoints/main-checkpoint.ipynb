# Fairness of scholastic aptitude test in New York City

In this project, we will aim to find how fair is the scholastic aptitude test (SAT) for the diverse student population in the schools of New York, US. The test consists of three section, each of which has a maximum score of 800. The combined maximum score is 2400. Since, the sat scores are used by colleges for admission of students, it is important that the test offers a fair chance to every student, irrespective of their race, gender, income and more. 

We will analyze the existing data about the sat test takers to judge its fairness. We will try to find out the following:

- the correlation between the average sat scores and the demographics of different schools in the dataset
- the correlation between the percentage of male and female students in a school versus the average sat scores of that school
- the correlation between the sat scores and other metrics such as the overall school district quality.

## Summary of the results

After analyzing the data, we found evidence that the SAT exams are unfair towards black and hispanic student population as compared to white and asian student population. Upon further investigation, we found that a possible reason for the poor performance of hispanic students could be the poor english skills. We also found that, the females perform better in the SAT exams than male student population.  

# Read in the data

We will use the datasets from the [NYC Open Data](https://data.cityofnewyork.us/Education) repository to perform our analysis. The repository contains the data on high school SAT scores, demographics on each high school and more. In particular, we will use the following data and surveys `ap_2010.csv`, `class_size.csv`, `demographics.csv`, `graduation.csv`, `hs_directory.csv`, `sat_results.csv`, `survey_all.txt`, and `survey_d75.txt`. 

```python
import pandas as pd
import numpy
import re

# Datasets to be read
data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

# Store the datasets in a dictionary
data = {}

for f in data_files:
    d = pd.read_csv("data/{0}".format(f))
    data[f.replace(".csv", "")] = d

# Display all the column without truncation
pd.set_option("display.max_columns", 500)

# Quick exploration of the data 
for key in data.keys():
    print(f"Dataset: {key}\n")
    print(data[key].head())
    print("\n")
```

# Read in the surveys

To read in the surveys, we have to specify the encoding, which is Windows-1252 for txt files. These txt files are tab delimited. 

```python
# Convert the survey from txt files to a pandas dataframe 
all_survey = pd.read_csv("data/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("data/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pd.concat([all_survey, d75_survey], axis=0) # Combine the two surveys

survey["DBN"] = survey["dbn"] # Copy the data to a new column

# Suvey columns which are relevant for our analysis
survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]

# Keep the subset of data relevant for our analysis
survey = survey.loc[:,survey_fields]
data["survey"] = survey # add the survey data to data dictionary

# Quick exploration of the survey data
data["survey"].head(5)
```

In the above cell, we only kept a subset of the survey data that is most relevant for our analysis. The keys to the `survey_fields` list is available in the [Survey Data Dictionary](https://data.cityofnewyork.us/Education/2011-NYC-School-Survey/mnz3-dyi8).

From the exploration above, we make the following observations
- Each dataset has a `DBN` column or the information to create one. This implies that we can use a `DBN` column to combine different datasets.
- Some of the datasets have multiple rows with the same `DBN` column. This means that we will have to remove or condense these rows to have a unique DBN value. 

# Add DBN columns

From the above analysis, we noticed that, datasets like `class_size` do not have a `DBN` column. However, from analysing the first few rows of the `sat_results` data, we observed that the `DBN` in `sat_results` data is just a combination of the `CSD` and `SCHOOL CODE` columns in the `class_size` data. Further, we noticed that the `DBN` is padded, so that the `CSD` portion always consists of two digits. Let's understand this better with the help of an example:

Suppose, `CSD` = 1, then `Padded CSD` = 01. However, if `CSD` = 19, then `Padded CSD` would be unchanged. Let us make the appropriate changes to obtain the `DBN` values.

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"] # Copy the "dbn" column to the new "DBN" column

# Add padding of 0 to single digit numbers in "CSD" column  
def pad_csd(num):
    string_representation = str(num)
    if len(string_representation) > 1:
        return string_representation
    else:
        return "0" + string_representation

# Store the padded CSD in a new column named 'padded_csd'
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)

# Combine the 'padded_csd' with 'SCHOOL CODE' to obtain the "DBN" in hs_directory dataset
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

We notice that the same school has multiple values in the `GRADE` and `PROGRAM TYPE` column, and that's why we have multiple rows with same `DBN` value. We choose to work with `09-12` value in the `GRADE` column, since we are dealing with high-schools.

We choose to work with `GEN ED` category, since it is the largest category by far. 

# Convert columns to numeric

To analyse the different correlations that we discussed in the introduction, we first need to convert the `SAT Math Avg. Score`, `SAT Critical Reading Avg. Score`, and `SAT Writing Avg. Score` columns in the `sat_results dataset` from string data type to numeric. Then, we can add these three columns to obtain the avg. sat scores.

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']

# Convert the data in cols to numeric datatype
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce") # error="coerce" converts invalid strings to missing values 
    
# Add the three numeric columns and store the result in the sat_score column of sat_results datasete
data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

# Extract the laitutde from the Location 1 column
def find_lat(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat

# Extract the longitude from the Location 1 column
def find_lon(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

# Store the extracted latitude and longitude of each school in the 'lat' and 'lon' columns of hs_directory
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

# Convert the latitude and longitude columns to numeric data type
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")
```

# Condense datasets

From the above analysis, we decided to work with `09-12` values in the `GRADE` column and `GEN ED` in the `PROGRAM TYPE` column. Doing so,  we hope to get a dataset which has unique `DBN` values.

```python
class_size = data["class_size"] # Copy the dataframe to a new variable

# Filter the dataset to choose only a subset
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]

# Quick exploration to verify the changes
class_size.head()
```

We notice that the `DBN` values are still not completely unique. This is due to the `CORE COURSE (MS CORE AND 9-12 ONLY) AND CORE SUBJECT (MS CORE AND 9-12 ONLY)`. Let us explore the `CORE SUBJECT (MS CORE AND 9-12 ONLY)` column further.

```python
# Unique values in the CORE SUBJECT (MS CORE and 9-12 ONLY) column
class_size["CORE SUBJECT (MS CORE and 9-12 ONLY)"].value_counts()
```

These values pertain to different subjects. We want to include all the subjects a school offers, so we can preserve the class size. We can do this by averaging across all of the classes a school offers. This way, we will obtain unique `DBN` values, while incorporating as much data as possible.

```python
# Group the data to have unique DBN values
class_size = class_size.groupby("DBN").agg(numpy.mean)
class_size.reset_index(inplace=True)
data["class_size"] = class_size # Assign the resulting dataset back to the data dictionary
```

### Condense the `demographics` and `graduation` datasets

The `demographics` dataset has multiple rows with the same `DBN` value. The `schoolyear` prevents a given `DBN` to be unique. We only select the rows with `20112012` entry. The `Demographic` and `Cohort` columns prevent a unique `DBN` in the `graduation` dataset. A `Cohort` refers to the year the data represents, and the `Demographic` refers to a specific demographic group. We pick the most recent `Cohort`, i.e. 2006, and `Total Cohort` `Demographic`.

```python
# Choose the subset of data with schoolyear = 20112012
data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]

# Choose the subset of data with Cohort as 2006 and Demographic as Total Cohort
data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]

# Quick check on the changes
data["graduation"].head(5)
```

# Convert AP scores to numeric

The [Advanced Placement](https://ap.collegeboard.org/#:~:text=AP%20gives%20students%20the%20chance,earn%20college%20credit%20and%20placement.)(AP) is a test, which high school students take before applying to college. These exams give students credits for college courses. AP exams have a 1 to 5 scale. AP exams are much more rare in schools with a lack of funding or academic rigor. 

We will analyse the correlation between the sat and AP scores. To do so, firstly we need to convert the following columns to numeric datatype
- AP Test Takers
- Total Exams Taken
- Number of Exams with scores 3 4 or 5.

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

# Convert cols to numeric datatype
for col in cols:
    data["ap_2010"][col] = pd.to_numeric(data["ap_2010"][col], errors="coerce")
```

# Combine the datasets

In this project, we are mainly concerned with determining demographic factors that correlate with sat scores, therefore, we want to preserve as many rows as possible from `sat_results` while minimizing null values. Firstly, we will "left" join `ap_2010` and `graduation` datasets to the `sat_results` dataset, since all the data in these datasets is very useful for our analysis. Then, we will merge `class_size`, `demographics`, `survey` and `hs_directory` using a "left" join. This way, we will only keep the data that is directly relevant for our analysis. After combining the data, we will deal with the null values that arise from performing the join operations.

```python
combined = data["sat_results"] # Assign sat_results dataset to the variable combined

# Merge the sat_result and the ap_2010 dataset 
combined = combined.merge(data["ap_2010"], on="DBN", how="left")

# Merge the combined with graduation dataset
combined = combined.merge(data["graduation"], on="DBN", how="left")

to_merge = ["class_size", "demographics", "survey", "hs_directory"]

# Merge the rest of the datasets to get a combined dataset
for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")
    
# Quick exploration of the combined dataset
combined.head(5)
```

```python
# Check the shape of the dataset after performing merge operation
print(f"Shape of combined dataset: {combined.shape}")
print(f"Shape of sat_results dataset: {data['sat_results'].shape}")
```

We observe that the inner join resulted in fewer rows than in the original `sat_result` dataset. 

```python
# Fill in the missing values using the mean
combined = combined.fillna(combined.mean())

# Fill the remaining missing values with 0
combined = combined.fillna(0)
```

# Perform analysis on the combined dataset

Finally, we have a clean set of data to perform our analysis. 

### Add a school district column 

The school district information could be helpful in analysisng the sat scores. So, we extract the information about the school district from the first two characters of the `DBN` values.

```python
# Extract the school district information
def get_first_two_chars(dbn):
    return dbn[0:2]

# Save the school district in a new column
combined["school_dist"] = combined["DBN"].apply(get_first_two_chars)
```

# Find correlations

In the cells below, we find the correlations between the `sat_score` column with all the other columns. We will consider correlations with correlation coefficient above 0.25 or below -0.25 as interesting. 

```python
correlations = combined.corr()

# Correlation between sat scores and other columns
correlations = correlations["sat_score"]

# Quick exploration of the correlation coefficients
print(correlations)
```

We observe that, the `SAT Critical Reading Avg. Score`, `SAT Math Avg. Score`, and `SAT Writing Avg. Score` are strongly correlated with the `sat_scores` column. This is unsurprising as the `sat_scores` are sum of these three columns. Further we notice the following 

- The `total_enrollement` has a strong positive correlation with the `sat_scores`.
- Both the number of females (`female_num`) and female percentage (`female_per`) correlate positively with `sat_scores`. On the other hand, the percentage of males (`male_per`) and the number of males (`male_num`) correlate negatively. This could indicate that women do better on the SAT than men.
- Teacher and student ratings of school safety (`saf_t_11`, and `saf_s_11`) correlate positively with `sat_score`.
- Student ratings of school academic standards (`aca_s_11`) correlate with sat_score, but this does not hold for ratings from teachers and parents (`aca_p_11` and `aca_t_11`).
- There is significant racial inequality in SAT scores (`white_per`, `asian_per`, `black_per`, `hispanic_per`).
- The percentage of English language learners at the school (`ell_percent`, `frl_percent`) has a strong negative correlation with SAT scores.

# Plot survey correlations

Next, we plot the correlations between the `sat_scores` and the columns of `survey_field` dataset.

```python
# Remove DBN since it's a unique identifier, not a useful numerical value for correlation.
survey_fields.remove("DBN")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Plot correlation between sat_scores and columns of survey field cols
plt.figure(figsize=(8,6))
correlations[survey_fields].plot.bar(color="#3DB2FF")
plt.plot(np.arange(-40,40), 0*np.arange(-40,40), color = "black", linewidth=0.7)
plt.title("Correlation between sat scores and survey_fields col", fontsize=16)
plt.ylabel("Correlation Index", fontsize=14)
plt.show()
```

We notice that the `saf_s_11`, `N_s`, `N_p` columns have strong positive correlation with the `sat_scores`. We will investigate the correlation between the `saf_s_11`, which refers to the safety rating of school as given by the students and the `sat_scores` further. 

```python
# Plot the student rating of school safety vs the sat scores
combined.plot.scatter(x="saf_s_11", y="sat_score", alpha = 0.6, color="#1E3163")
plt.title("Sat Scores vs Security perceived by students", fontsize=14)
plt.xlabel("security score", fontsize=14)
plt.ylabel("sat scores", fontsize=14)
plt.show()
```

From the plot above, we notice that the `sat_scores` have a strong positive correlation with the student rating of school safety. This implies that, students who feel more secure at their school tend to perform better in their SATs. It will be interesting to explore the relationship between the average safety rating by students in each borough and the corresponding average sat scores in each borough. 

```python
avg_safety_score = combined.groupby("boro").agg(np.mean)["saf_s_11"]

avg_safety_score.plot.bar(rot=30, color="#3DB2FF")
plt.title("Avg saftey in each borough", fontsize=16)
plt.xlabel("Borough", fontsize=14)
plt.ylabel("Average Safety Score", fontsize=14)
plt.show()
```

# Racial differences in SAT scores

In this section, we will investigate the correlation between the percentage of, white students (`white_per`), asian students (`asian_per`), black students (`black_per`), hispanic students (`hispanic_per`) and the sat scores.

```python
# Plot the correlation index 
x_ticks_per = ["white %", "asian %", "black %", "hispanic %"]
correlations[["white_per", "asian_per", 
              "black_per", "hispanic_per"]].plot.bar(rot=0, color="#93B5C6")
plt.title("Correlation between sat scores and \n percentage of racial population", fontsize=14)
plt.ylabel("Correlation Index", fontsize=13)
plt.xticks([0,1,2,3], x_ticks_per, fontsize=13)
plt.plot(np.arange(-10,10), 0*np.arange(-10,10), color="black", linewidth=0.6)
plt.show()
```

The above analysis indicates that, the SAT scores are negatively correlated with percentage of black and hispanic students in the school. But, the SAT scores have a strong positive correlation with the percentage of white and asian students. This implies that the black and hispanic students tend to do worse in SAT exams as opposed to the white and asian students. Among the black and hispanic students, the later seem to perform worse in the SAT.

## Further Investigation of Performance of Hispanic Students in SAT

As we saw above, the hispanic students seemed to perform the worst in SAT. We will further analyse this in the cells below.

```python
# Plot the percentage of hispanics in a school vs avg. sat scores
combined.plot.scatter(x="hispanic_per", y="sat_score", color="#1E3163", alpha=0.5)
plt.title("Sat scores vs % of hispanics in school", fontsize=15)
plt.xlabel("percentage of hispanics", fontsize=13)
plt.ylabel("avg. sat score", fontsize=13)
plt.show()
```

From the above scatter plot, we can infer that schools which higher percentage of hispanic students tend to do worse in SAT than the schools which have lower percentage of hispanic students. To investigate the reason of the lower SAT scores among the schools with higher Hispanic population, we select the schools which have higher than 95% of hispanic population. 

```python
# Schools with greater than 95% hispanic student population
combined[combined["hispanic_per"] > 0.95].head(5)
```

There is an interesting column `ell_percent` in the above dataframe, which refers to the percentage of english language learners in a school. Since, the SAT scores take into consideration the english literacy of a student, we contemplate that the schools with higher percentage of hispanic population have a higher percentage of english language learners. 

```python
# Percentage of english learners in schools with > 95% hispanic students. 
combined[combined["hispanic_per"] > 0.95]["ell_percent"].plot.hist(color="#93B5C6")
plt.xlabel("Percentage of english language learners")
plt.show()
```

We observe that, a number of these schools have a significant student population which are english language learners. Further, let us analyze the schools which have less than ten percent hispanic student population and greater than 1800 SAT scores. 

```python
# Schools with less hispanic students and above avg. SAT scores
combined[(combined["hispanic_per"] < 10) & (combined["sat_score"] > 1800)]
```

```python
# The percentage of english language learners in the above schools
combined[(combined["hispanic_per"] < 10) & (combined["sat_score"] > 1800)]["ell_percent"].value_counts()
```

We observe that, schools with a lower percentage of Hispanic students and above average SAT scores, are the ones which have less than 0.2 percent of english language learners. This indicates that SAT could be unfair to the Hispanic students because they tend to have weaker english language skills.

Next, we want to explore the correlation between the percentage of AP test takers and the sat scores in a school. Since, only a few selective students appear for the AP exams, which gives them college credit, we contemplate if it is positively correlated with the sat scores. 

```python
# Calcute the percentage of students in each school that took an AP exam
combined["ap_per"] = combined["AP Test Takers "] / combined["total_enrollment"]
```

```python
# Plot ap_per vs sat_score
combined.plot.scatter(x="ap_per", y="sat_score", color="#7D1935", alpha=0.4)
plt.title("Sat scores vs % of students who took \nAP in a school", fontsize=15)
plt.xlabel("% of AP test takers", fontsize=13)
plt.ylabel("sat scores", fontsize=13)
plt.show()
```

As opposed to our earlier hypothesis, we see no clear trend that can help us conclude that the higher percentage of AP test takers are positively correlated with higher sat scores. 

We conclude our analysis by observing a few points
- The SAT scores are positively correlated with the students rating of safety at the school. 
- The SAT scores are negatively correlated with the percentage of blacks and hispanics at a school. This implies that the SAT is unfair towards blacks and hispanic students. Moreover, we note that the hispanics tend to do worse than blacks in the SAT. Our analysis further revealed that, this could be because of lower english literacy among the hispanics. 
- We further found that, one average the female students tend to perform better in the SAT than the male students. 
- Lastly, we found that, the percentage of AP test takers does not have a strong correlation with the sat scores.
